```
def stats_sentence_list(sen_list, is_print=False, need_filter=True):
  filter_verbs = ['是', '有', '没有', '不是']
  v_list = dict()
  for sentence in sen_list[:]:
    words = pseg.cut(sentence)

    for word, flag in words:
        if flag.startswith('v') and (word not in filter_verbs or not need_filter):
          if is_print:
            print('%s, %s' % (word, flag))
          if word not in v_list:
            v_list[word] = 1
          else:
            v_list[word] = v_list[word] + 1
    if is_print:
      print('-----------------')
  return v_list
  
```

```
def count_portion(verb_dict, first_count):
  verb_freq_sum = sum(verb_dict.values())
  sorted_v_list = sorted(verb_dict.items(), key=lambda kv: kv[1], reverse=True)
  first_v_freq_sum = sum([x[1] for x in sorted_v_list[:first_count]])
  return first_v_freq_sum/verb_freq_sum
```

```
def mean_length(sen_list):
  sum_char = sum([len(sen) for sen in sen_list])
  return sum_char/len(sen_list)
```
