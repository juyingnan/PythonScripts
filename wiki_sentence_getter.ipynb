```
zh_lans = [lang for lang in wikipedia.languages() if lang.startswith('zh')]
zh_lans
```

```
# wikipedia.languages()
```

```
import random

lang = zh_lans[0]
wikipedia.set_lang(lang)

count = 200
sentence_list = list()
while count > 0:
  try:
      topic = wikipedia.random(pages=1)
  except:
      continue
      
  try:
    page = wikipedia.page(topic)
  except:
    continue
  
  sentences = page.summary.split('。')
  if len(sentences) > 0:
    for i in range(20):
      random_sentence = random.choice(sentences).strip()
      if len(random_sentence) > 10 and len(random_sentence) < 30 and not random_sentence.startswith('=') and '\n' not in random_sentence:
        sentence_list.append(random_sentence)
        count -= 1
        break
  
sentence_list
```

```
for sentence in sentence_list:
  print(sentence)
```
